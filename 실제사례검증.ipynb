```python
import pandas as pd
```

```python
df_adjacency = pd.read_csv('drug-SE adjacency matrix.csv', index_col=0)
df_adjacency
```

```python
drug_list = df_adjacency.index
drug_list
```

```python
df_score = pd.read_csv('score_matrix.csv', index_col=0)

# drug_list를 df_score의 새로운 컬럼으로 추가
df_score[''] = drug_list

#drug를 인덱스로 설정
df_score = df_score.set_index('')

df_score
```

```python
# df_score 히트맵 생성
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(7, 7))
sns.heatmap(df_score, cmap='crest', annot=False)

plt.show()
```
